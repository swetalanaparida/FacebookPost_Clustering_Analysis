**OUTLIER DETECTION AND TREATMENT**

**LOCAL OUTLIER FACTOR(LOF)**
-> It assesses how isolated each data point is compared to its neighbors. It calculates a score for each post based on how differently engaged it is relative to nearby posts. If a post's score is much higher than its neighbors, LOF flags it as an outlier. The contamination parameter helps LOF decide how many posts to consider as outliers based on their engagement patterns compared to others.


---
**KEY POINTS-**

->**Contamination Parameter :** Adjust contamination to control the threshold for outliers. Higher values allow more outliers to be detected.

->**Output :**LOF outputs -1 for outliers and 1 for inliers. Counting -1s gives the number of outliers.

```python
from sklearn.neighbors import LocalOutlierFactor

# Numeric_cols are defined as-
numeric_cols = ['num_reactions', 'num_comments', 'num_shares', 'num_likes', 'num_loves',
                'num_wows', 'num_hahas', 'num_sads', 'num_angrys']

# Creating an instance of LOF with contamination set to 0.05
# Means you accept 5% of the data points to be outliers
clf_lof = LocalOutlierFactor(contamination=0.05)

# Fit the LOF model and predict outliers
outliers_lof = clf_lof.fit_predict(df[numeric_cols])

# Counting the number of outliers
num_outliers_lof = (outliers_lof == -1).sum()

# Selecting non-outliers and copying them to a new DataFrame
df_cleaned = df[outliers_lof == 1].copy()

# Displaying information about the cleaned dataset
print("Information about the dataset after removing outliers:")
print(df_cleaned.info())

# Printing the number of outliers detected
print(f"\nNumber of outliers detected by LOF: {num_outliers_lof}")

# Printing the percentage of outliers detected
print(f"\nPercentage of outliers detected by LOF: {num_outliers_lof / len(df) * 100:.2f}%")

# Printing number of entries after removing outliers
print(f"\nNumber of entries after removing outliers: {len(df_cleaned)}")
```

**Results of Outlier Detection:**

**Number of Outliers Detected:** A total of 353 outliers were identified.

**Percentage of Outliers:** This represents 5.01% of the total data.

**Data After Removing Outliers:** The cleaned dataset now has 6,697 entries, down from the original 7,050.

# **CLUSTER MODELLING :**

**Clustering** is a technique used in data analysis to group similar data points together. Here are some common clustering methods:

**A) K-Means Clustering:**
Divides data into k distinct, non-overlapping subsets (clusters) based on distance from the centroid of each cluster.


**B) DBSCAN (Density-Based Spatial Clustering of Applications with Noise):**
Groups together points that are closely packed while marking points in low-density regions as outliers.


**C) Agglomerative Clustering:**
A type of hierarchical clustering that starts with each data point as a single cluster and merges the closest pairs of clusters iteratively.


**A) K-MEANS METHOD OF CLUSTERING**

**DETERMINE OPTIMAL NUMBER OF CLUSTERS :**

**ELBOW METHOD**

```python
# Now lets find out the optimal number of clusters
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
wcss=[]
for i in range(2,10):
  kmeans=KMeans(n_clusters=i,n_init=25,random_state=1234)
  kmeans.fit(df_scaled)
  wcss.append(kmeans.inertia_)

plt.plot(range(2,10),wcss,marker='o', linestyle='--')
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.xticks(range(0,10))
plt.show()
```

Taking optimum k as 3 from elbow method on WCSS.

**Observation-**
 - On the x-axis,the number of clusters (k), ranging from 0 to 10.
 - On the y-axis, there is "Within Cluster Sum of Squares" (WSS),
which tells us how tightly packed the points in each cluster are.

Asthe number of clusters increases, the WSS decreases, means that the points in each cluster are closer together.
However, beyond a certain point (around 3 clusters here), the rate of improvement slows down significantly.
This point looks like an elbow in the graph.

The elbow at k=3 suggests that 3 is a good number of clusters for this data.
Beyond this point, adding more clusters doesn't significantly improve how well the points are grouped.

**FINDING THE OPTIMAL NUMBER OF CLUSTERS**

**SILHOUETTE SCORE :**

```python
from sklearn.metrics import silhouette_score

# Initialize an empty list to store silhouette scores
silhouette_scores = []

# Calculate silhouette scores for different numbers of clusters
for k in range(2,10):
    kmeans = KMeans(n_clusters=k, random_state=1234, n_init=25)
    cluster_labels = kmeans.fit_predict(df_scaled)
    silhouette_avg = silhouette_score(df_scaled, cluster_labels)
    silhouette_scores.append(silhouette_avg)

# Plotting Silhouette scores
import matplotlib.pyplot as plt

# plt.figure(figsize=(10, 8))
plt.plot(range(2, 10), silhouette_scores, marker='o', linestyle='--')
plt.title('Silhouette Method for Optimal k')
plt.xlabel('Number of Clusters(k)')
plt.ylabel('Silhouette Score')
plt.grid(True)
plt.show()
```

Taking optimum k as 2 from silhouette method.

**Observations-**

 x-axis: Number of groups (clusters), ranging from 2 to 9.

 y-axis: Average score showing how well the points fit into their groups (higher is better).

The curve reaches its highest point when the number of clusters is 2.
This suggests that dividing the data into 2 clusters results in the most well-defined and distinct clusters.

**CALINKSKI - HARABASZ SCORE :**

```python
from sklearn.metrics import calinski_harabasz_score

# List to store Calinski-Harabasz scores for different number of clusters
calinski_harabasz_scores = []

# Range of cluster numbers to evaluate
cluster_range = range(2, 10)

# Calculate the score for each number of clusters
for k in cluster_range:
    kmeans = KMeans(n_clusters=k, random_state=1234, n_init=25)
    labels = kmeans.fit_predict(df_scaled)
    score = calinski_harabasz_score(df_scaled, labels)
    calinski_harabasz_scores.append(score)

# Plotting the Calinski-Harabasz scores
# plt.figure(figsize=(10, 8))
plt.plot(cluster_range, calinski_harabasz_scores, marker='o', linestyle='--')
plt.title('Calinski-Harabasz Score for Optimal k')
plt.xlabel('Number of Clusters(k)')
plt.ylabel('Calinski-Harabasz Score')
plt.show()
```

Taking optimum k as 3 from calinski harabasz method.

**Observations-**

x-axis: Number of clusters (k), ranging from 2 to 9.

y-axis: Calinski-Harabasz score, which measures cluster separation and compactness.

Higher values indicate better-defined clusters.
The score reaches its highest point when the data is divided into 3 clusters.
This suggests that dividing the data into 3 clusters results in clusters that are well-separated and compact.

**B) DBSCAN METHOD OF CLUSTERING :**

```python
# Define the numerical features
numeric_cols = ['num_reactions', 'num_comments', 'num_shares', 'num_likes', 'num_loves',
                'num_wows', 'num_hahas', 'num_sads', 'num_angrys']
X = df_cleaned[numeric_cols]
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
df_scaled = pd.DataFrame(X_scaled, columns=numeric_cols)

# Display the scaled data
print("Scaled Data:")
print(df_scaled)
```

**Data Scaling:**
The data has been scaled so that each feature (like number of reactions, comments, likes) has a mean of 0 and a standard deviation of 1. This means the data is now on a common scale, making it easier to compare and analyze.

```python
from sklearn.neighbors import NearestNeighbors
import matplotlib.pyplot as plt

# Find the optimal value of 'eps'
neighbors = NearestNeighbors(n_neighbors=4)
neighbors_fit = neighbors.fit(X_scaled)
distances, indices = neighbors_fit.kneighbors(X_scaled)

# Sort the distances
distances = np.sort(distances, axis=0)
distances = distances[:, 1]

plt.plot(distances)
plt.title('K-distance Graph')
plt.xlabel('Data Points sorted by distance')
plt.ylabel('Epsilon')

#draw a line at the choosen value of eps
plt.axhline(y=0.7, color='r', linestyle='--')
```

**Elbow Point:** Around the 6000th data point, there is a noticeable sharp increase or "elbow" in the curve.
This point suggests a threshold where the distance between data points starts to increase significantly.

**Optimal Epsilon:** The sharp increase point, also known as the "elbow", is typically considered the optimal value for epsilon.
In this graph, the elbow occurs at an epsilon value slightly above 0.7.
Choosing optimal epsilon as 0.7.

**USING SILHOUETTE SCORE FOR FINDING THE VALUE OF MIN SAMPLES**

```python
from sklearn.metrics import silhouette_score
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
import seaborn as sns

ss = []

for ms in range(1,15):
    dbscan = DBSCAN(eps=0.7, min_samples=ms)
    dbscan.fit(X_scaled)
    silhouette = silhouette_score(X_scaled, dbscan.labels_)
    ss.append(silhouette)

# Creating a Series for silhouette scores
ss_series = pd.Series(ss, index=range(1,15))

# Plotting silhouette scores
plt.figure(figsize=(10, 6))
sns.lineplot(x=ss_series.index, y=ss_series.values, marker='o')
plt.title('Silhouette Score for Different min_samples Values')
plt.xlabel('Number of DBSCAN points in neighborhood (min_samples)')
plt.ylabel('Silhouette Score')
plt.xticks(range(1,15))
plt.grid(True)
plt.show()
```

```python
 from sklearn.metrics import silhouette_score

# Assuming the optimal 'eps' found is 0.7 (example)
optimal_eps = 0.7
min_samples_range = range(2, 10)
best_score = -1
best_min_samples = None

for min_samples in min_samples_range:
    dbscan = DBSCAN(eps=optimal_eps, min_samples=min_samples)
    clusters = dbscan.fit_predict(X_scaled)

    if len(set(clusters)) > 1:  # Only consider if more than 1 cluster is formed
        score = silhouette_score(X_scaled, clusters)
        if score > best_score:
            best_score = score
            best_min_samples = min_samples

print(f"Best silhouette score: {best_score} with min_samples: {best_min_samples}")
```

**Silhouette Score:** This score measures how well the clusters are defined. A higher score means that the clusters are well-separated and distinct.

**Best Score:** The highest silhouette score obtained was approximately 0.30. This is a measure of how well the clusters formed by the DBSCAN algorithm distinguish between different groups.

**Optimal Min Samples:** The best clustering result was achieved with 7 minimum samples required to form a cluster. This means that in the DBSCAN algorithm, a group of 7 or more data points is needed to be considered a valid cluster.

```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
import seaborn as sns

# Trying different numbers of clusters to find the optimal one
sil_scores = []
range_n_clusters = range(2, 10)

for n_clusters in range_n_clusters:
    agglom = AgglomerativeClustering(n_clusters=n_clusters)
    cluster_labels = agglom.fit_predict(X_scaled)

    # Calculate the silhouette score for the current number of clusters
    sil_score = silhouette_score(X_scaled, cluster_labels)
    sil_scores.append(sil_score)

# Creating a Series for silhouette scores
sil_series = pd.Series(sil_scores, index=range_n_clusters)

# Plotting silhouette scores
plt.figure(figsize=(10, 6))
sns.lineplot(x=sil_series.index, y=sil_series.values, marker='o')
plt.title('Silhouette Score for Different Numbers of Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.xticks(range_n_clusters)
plt.grid(True)
plt.show()

# Choose the optimal number of clusters based on the silhouette scores
optimal_n_clusters = sil_series.idxmax()
print(f"Optimal number of clusters: {optimal_n_clusters}")
```

**Optimal Number of Clusters:** The analysis suggests that 2 clusters is the best choice. This is based on the silhouette scores, which measure how well the data points fit within their clusters compared to others. A higher score indicates better-defined clusters.

 For Facebook posts data, dividing it into 2 main clusters gives the clearest and most meaningful separation, making it the most suitable choice for understanding the patterns in your data.

```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import calinski_harabasz_score
import matplotlib.pyplot as plt
import numpy as np

# Range for number of clusters
range_n_clusters = range(2, 10)

# Arrays to store scores

calinski_scores = []

for n_clusters in range_n_clusters:
    agglom = AgglomerativeClustering(n_clusters=n_clusters)
    cluster_labels = agglom.fit_predict(X_scaled)

    # Calculate Calinski-Harabasz Index
    calinski_scores.append(calinski_harabasz_score(X_scaled, cluster_labels))

# Calinski-Harabasz Index plot
plt.figure(figsize=(10, 6))
plt.plot(range_n_clusters, calinski_scores, marker='o')
plt.title('Calinski-Harabasz Index for Optimal Number of Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('Calinski-Harabasz Score')
plt.grid(True)
plt.show()

# Determine the optimal number of clusters based on the highest Calinski-Harabasz score
optimal_n_clusters_calinski = range_n_clusters[np.argmax(calinski_scores)]
print(f"Optimal number of clusters (Calinski-Harabasz Index): {optimal_n_clusters_calinski}")
```

For Facebook posts data, splitting it into 3 clusters will give you the most meaningful separation, where each cluster is well-defined and distinct from the others.

**COMPARING THESE THREE METHODS TO FIND OUT WHICH ONE IS BEST.**

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans, DBSCAN, AgglomerativeClustering
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score

# Dropping non-numeric columns and filling missing values
df_cleaned = df.select_dtypes(include=[np.number]).dropna()

# Standardizing the data
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_cleaned)

# K-Means Clustering
kmeans = KMeans(n_clusters=3,n_init=25, random_state=1234)
df_cleaned['Cluster'] = kmeans.fit_predict(df_scaled)

# DBSCAN Clustering
dbscan = DBSCAN(eps=0.7, min_samples=7)
df_cleaned['Cluster_DBSCAN'] = dbscan.fit_predict(df_scaled)

# Agglomerative Clustering
agg = AgglomerativeClustering(n_clusters=3)
df_cleaned['Cluster_Agg'] = agg.fit_predict(df_scaled)

# Function to evaluate clustering models
def evaluate_clustering_models(df, df_scaled):
    results = {}

    # K-Means
    kmeans_labels = df['Cluster']
    results['K-Means'] = {
        'Silhouette Score': silhouette_score(df_scaled, kmeans_labels),
        'Calinski-Harabasz Index': calinski_harabasz_score(df_scaled, kmeans_labels),
        'Davies-Bouldin Index': davies_bouldin_score(df_scaled, kmeans_labels)
    }

    # DBSCAN
    dbscan_labels = df['Cluster_DBSCAN']
    if len(set(dbscan_labels)) > 1:  # Ensure DBSCAN produced more than one cluster
        results['DBSCAN'] = {
            'Silhouette Score': silhouette_score(df_scaled, dbscan_labels),
            'Calinski-Harabasz Index': calinski_harabasz_score(df_scaled, dbscan_labels),
            'Davies-Bouldin Index': davies_bouldin_score(df_scaled, dbscan_labels)
        }
    else:
        results['DBSCAN'] = {
            'Silhouette Score': float('-inf'),
            'Calinski-Harabasz Index': float('-inf'),
            'Davies-Bouldin Index': float('inf')
        }

    # Agglomerative Clustering
    agg_labels = df['Cluster_Agg']
    results['Agglomerative'] = {
        'Silhouette Score': silhouette_score(df_scaled, agg_labels),
        'Calinski-Harabasz Index': calinski_harabasz_score(df_scaled, agg_labels),
        'Davies-Bouldin Index': davies_bouldin_score(df_scaled, agg_labels)
    }

    return results

# Evaluating clustering models
results = evaluate_clustering_models(df_cleaned, df_scaled)

# Displaying the results
for model, metrics in results.items():
    print(f"\n{model} Clustering:")
    for metric, score in metrics.items():
        print(f"{metric}: {score:.4f}")
```
